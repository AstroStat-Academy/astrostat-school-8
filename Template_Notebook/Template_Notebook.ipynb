We study the **mean** as a central summary of data.<br>
It represents the typical value in a dataset and gives a quick sense of where observations are concentrated.

```python
import numpy as np
np.mean([1, 2, 3, 4, 5])
```

The **variance** measures how spread out values are around the mean.<br>
It captures the average squared deviation, so larger values indicate more dispersion.

> _NOTE: Variance is sensitive to extreme values and can be inflated by outliers._

The **standard deviation** is the square root of variance.<br>
It provides a more interpretable measure of spread since it is expressed in the same units as the data.

$\rightarrow$ The latter two are strongly connected.

<div style="border-left: 4px solid #2196F3; padding-left: 1em; padding-top: 1em; padding-bottom: 1em; margin: 1em 0; background: #f0f8ff; color: #1a1a1a;">

<details>

<b><summary>[More details] (click here to expand)</summary></b>  

The mean is computed as the sum of all observations divided by their number.  
Variance quantifies dispersion by averaging squared distances from the mean.  
Standard deviation rescales this quantity, making it easier to compare with the original data.

</details>
</div>
